## Introduction

This notebook demonstrates an implementation of the technique described in 
"*A Rigorous Technique for Measuring the Scattering Matrix of a Multiport Device with a 2-Port Network Analyzer*" [1]. The basic concept is to measure series n-port sub-network, and then compile the results to form the full n-port. Generally, this is done by simply terminating each non-measured port with a matched load, and assuming the reflected power is negligable. For some environments this may not provide the most accurate possible results, or even be possible. 



[1] J. C. Tippet and R. A. Speciale, “A Rigorous Technique for Measuring the Scattering Matrix of a Multiport Device with a 2-Port Network Analyzer,” IEEE Transactions on Microwave Theory and Techniques, vol. 30, no. 5, pp. 661–666, May 1982.


```python
import skrf as rf 
from itertools import combinations

wg = rf.wr10
wg.frequency.npoints = 101
four_port = wg.random(n_ports  = 4)
four_port
```

```python
loads = [wg.load(.4),
         wg.load(.2+.3j),
         wg.load(.1-.2j),
         wg.load(.5j),
         ]
z_loads = array([k.z.flatten() for k in loads]).T
```

```python
composite = wg.match(nports = 4)

ports = arange(four_port.nports)

return_loss_port1 = []
return_loss_port1_renorm = []
measured,measured_renorm = {},{}


print 'a b c d e'
# ports  `a` and `b` are the ports we will connect the VNA too
for a,b in port_combos:
    # port `c` and `d` are the ports which we will connect the loads too
    c,d =ports[(ports!=a)& (ports!=b)] 
    
    # determine where `d` will be on four_port, after its reduced to a three_port
    e = where(ports[ports!=c]==d)[0][0]
    
    print a,b,c,d,e
    
    three_port = rf.connect(four_port,c, loads[c],0)
    two_port =  rf.connect(three_port,e, loads[d],0)
    
    
    measured['%i%i'%(a,b)] = two_port.copy()
    two_port.renormalize(c_[z_loads[:,a],z_loads[:,b]])
    measured_renorm['%i%i'%(a,b)] = two_port
    
    for i,m in enumerate([a,b]):
        for j,n in enumerate([a,b]):
            composite.s[:,m,n] = two_port.s[:,i,j]
    
    #if a == 0:
    #    return_loss_port1.append(two_port.s11)
    #    two_port.renormalize(z0_loads[:,:2])
    #    return_loss_port1_renorm.append(two_port.s11)
            
composite.z0 = z_loads
composite.renormalize(50)
```
