# The *new* **Network** Class

This notebook demonstrates the new design of skrf's **Network** Class. The new class utilizes a more object-oriented approach which is cleaner and more scalable. The draw-back is that it breaks backward compatibility.

## Intro

Creating a *new style* **Network** from an old

```python
from skrf import network2
reload(network2)

a = network2.Network.from_old_Network(rf.data.ring_slot)
```

There are two new sub-classes which to **Network**, 

* **Parameter** (like s,z,y,etc)
* **Projection** ( like db10, db20, deg, mag,etc)
 
 Accessing the network parameters returns on object, not an array as before

```python
print a.s
```

However, you can get at the array by accessing `val`. I think this is kind of annoying, but it allows for a more coherent design, as we will see below. 

```python
a.s.val[:5]
```

One reason for doing this is that plot functions become members of the parameters, which behave like you expect

```python
a.s.plot()
```

Same thing for other parameters,

```python
a.z.plot()
```

A further advantage is that each parameter has members which represent various scalar projections. 

```python
print a.s.db10
```

```python
a.s.db10.val[:5]
```

These also have plotting functions as you expect

```python
a.s.db10.plot()
```

```python
a.s.deg.plot()
```
